```python
 import numpy as np # linear algebra
 import pandas as pd
 import matplotlib.pyplot as plt
 import seaborn as sns
```

```python
df = pd.read_csv('/content/kidney_disease.csv')
```

```python
df.sample(5)
```

```python
df.columns
```

```python
df=df.iloc[:,[1,2,-1]]
```

### Train test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop('classification',axis=1),df['classification'],test_size=0.3,random_state=0)
```

```python
x_train.shape,x_test.shape
```

### StandardScaler

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# fit the scaler to the train set , it will learn the parameter
scaler.fit(x_train)

# transform train snd test sets
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
scaler.mean_
```

```python
x_train_scaled = pd.DataFrame(x_train_scaled, columns= x_train.columns)
x_test_scaled = pd.DataFrame(x_test_scaled, columns= x_test.columns)
```
